```python
import json
file_name = "../data/ave_pos_score_dict.json"
with open(file_name,'r',encoding='utf-8') as f:
    ave_pos_score_dict = json.load(f)
sa2_maincode_2016 = ave_pos_score_dict.keys()
pos_score_list = []
for i in sa2_maincode_2016:
    pos_score_list.append(ave_pos_score_dict[i])
from pandas.core.frame import DataFrame
c={"sa2_maincode_2016" : sa2_maincode_2016,
   "pos_score" : pos_score_list}
pos_score=DataFrame(c)
print(pos_score)
```

```python
import json
file_name = "../data/ave_compound_score_dic.json"
with open(file_name,'r',encoding='utf-8') as f:
    ave_compound_score_dict = json.load(f)
sa2_maincode_2016 = ave_compound_score_dict.keys()
compound_score_list = []
for i in sa2_maincode_2016:
    compound_score_list.append(ave_compound_score_dict[i])
from pandas.core.frame import DataFrame
c={"sa2_maincode_2016" : sa2_maincode_2016,
   "compound_score" : compound_score_list}
compound_score=DataFrame(c)
print(compound_score)    
```

```python
import json
file_name = "../data/ave_neu_score_dict.json"
with open(file_name,'r',encoding='utf-8') as f:
    ave_neu_score_dict = json.load(f)
sa2_maincode_2016 = ave_neu_score_dict.keys()
neu_score_list = []
for i in sa2_maincode_2016:
    neu_score_list.append(ave_neu_score_dict[i])
from pandas.core.frame import DataFrame
c={"sa2_maincode_2016" : sa2_maincode_2016,
   "neu_score" : neu_score_list}
neu_score=DataFrame(c)
print(neu_score)  
```

```python
import json
file_name = "../data/ave_neg_score_dict.json"
with open(file_name,'r',encoding='utf-8') as f:
    ave_neg_score_dict = json.load(f)
sa2_maincode_2016 = ave_neu_score_dict.keys()
neg_score_list = []
for i in sa2_maincode_2016:
    neg_score_list.append(ave_neg_score_dict[i])
from pandas.core.frame import DataFrame
c={"sa2_maincode_2016" : sa2_maincode_2016,
   "compound_score" : compound_score_list,
   "pos_score" : pos_score_list,
   "neu_score" : neg_score_list,
  "neg_score" : neg_score_list}
all_score=DataFrame(c)
print(all_score)  
```
